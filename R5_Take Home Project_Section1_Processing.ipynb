```python
## Importing the allowed libraries

import pandas as pd
import json
```

```python
## Accessing the JSON file on Google Drive

json_file_path = '/content/drive/My Drive/Python/R5/taylor_swift_spotify.json'
```

```python
## Loading the JSON file into the Colab environment

with open(json_file_path, 'r') as file:
    data = json.load(file)

# Now, 'data' contains the content of the JSON file
```

```python
# Assuming 'data' is a dictionary loaded from the JSON file
df = pd.DataFrame.from_dict(data)

# Now 'df' is a pandas DataFrame containing the JSON data
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', -1)
```

```python
df.head(2)
```

```python
## Flatten the JSON structure and create a DataFrame
df = pd.json_normalize(
    data,
    record_path=['albums', 'tracks'],
    meta=[
        'artist_id',
        'artist_name',
        'artist_popularity',
        ['albums', 'album_id'],
        ['albums', 'album_name'],
        ['albums', 'album_release_date'],
        ['albums', 'album_total_tracks']
    ],
    errors='ignore'
)

## Renaming columns to match the desired format
df.rename(columns={
    'track_audio_features.danceability': 'audio_features.danceability',
    'track_audio_features.energy': 'audio_features.energy',
    'track_audio_features.key': 'audio_features.key',
    'track_audio_features.loudness': 'audio_features.loudness',
    'track_audio_features.mode': 'audio_features.mode',
    'track_audio_features.speechiness': 'audio_features.speechiness',
    'track_audio_features.acousticness': 'audio_features.acousticness',
    'track_audio_features.instrumentalness': 'audio_features.instrumentalness',
    'track_audio_features.liveness': 'audio_features.liveness',
    'track_audio_features.valence': 'audio_features.valence',
    'track_audio_features.tempo': 'audio_features.tempo',
    'track_audio_features.id': 'audio_features.id',
    'track_audio_features.time_signature': 'audio_features.time_signature',
    'albums.album_id':'album_id',
    'albums.album_name':'album_name',
    'albums.album_release_date':'album_release_date',
    'albums.album_total_tracks':'album_total_tracks'
}, inplace=True)
```

```python
# Path to the CSV file
file_path = '/content/drive/My Drive/Python/R5/dataset.csv'

# Read the CSV file into a DataFrame named 'dataset'
dataset = pd.read_csv(file_path)

dataset.head(3)
```

```python
## Comparing the two DataFrames

# 1. Shape Comparison
shape_match = df.shape == dataset.shape

# 2. Column Names Comparison
columns_match = df.columns.tolist() == dataset.columns.tolist()

# 3. Data Types Comparison
dtypes_match = df.dtypes.equals(dataset.dtypes)

# 4. Data Comparison (this can be computationally intensive for large DataFrames)
data_match = df.equals(dataset)

# 5. Descriptive Statistics Comparison
stats_match = df.describe().equals(dataset.describe())

# Print the results
print(f'Shape match: {shape_match}')
print(f'Column names match: {columns_match}')
print(f'Data types match: {dtypes_match}')
print(f'Data match: {data_match}')
print(f'Descriptive statistics match: {stats_match}')
```

```python
## Comparing data types of each column in both DataFrames and only report mismatches
for col in dataset.columns:
    if dataset[col].dtype != df[col].dtype:
        print(f"Column: {col}")
        print(f"  dataset type: {dataset[col].dtype}")
        print(f"  df type: {df[col].dtype}")
        print(f"  Match: {dataset[col].dtype == df[col].dtype}")
        print()
```

```python
## Convert float columns to float64
float_cols = ['audio_features.danceability', 'audio_features.acousticness', 'audio_features.tempo']
for col in float_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

## Convert int columns to int64
int_cols = ['artist_popularity']
for col in int_cols:
    ## First convert to float to handle NaNs and then to int, because NaN cannot exist in an int type column
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype('int64')

## Display the new data types for the modified columns
print(df[float_cols + int_cols].dtypes)
```

After reviewing the original files as plain text, it looks like the boolean values in *dataset.csv* are: True, False. On the other hand, it looks like the boolean values in *taylor_swift_spotify.json* are: true, false. Having that in mind...

```python
# Convert the 'explicit' column from boolean to string with capitalized "True" and "False"
df['explicit'] = df['explicit'].map({True: 'True', False: 'False'})
dataset['explicit'] = dataset['explicit'].map({True: 'True', False: 'False'})

# Now the 'explicit' column in df should match the format in the dataset DataFrame
try:
    pd.testing.assert_frame_equal(df, dataset)
except AssertionError as e:
    print(e)
```
